```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load the image
image = cv2.imread('images/lenna.png')

# Define a 3x3 averageing kernel

kernel = np.ones((3,3), np.float32)/9

#Apply the kernel to smooth the image
smoothed_img = cv2.filter2D(image , -1, kernel)
```

```python
# Show the smoothend image
cv2.imshow('Original Image',image)
cv2.imshow('Smoothend Image',smoothed_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Load the image
gray = cv2.imread('images/lenna.png', 0)
kernel_size = 5# Adjust as needed
kernel = cv2.getGaussianKernel(kernel_size,1.25)
blurred =  cv2.filter2D(gray, -1, kernel) #Apply filter
```

```python
# Show the smoothend image
cv2.imshow('Original Image',gray)
cv2.imshow('Blured Image',blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
