```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/mtcars.csv')
```

```python
###### Seperate the input and output Variable
```

```python

#disp, hp, wt (We take this as input[[]])
x = df[['disp','hp','wt']]

# take mpg as output[]
y = df['mpg']
```

```python
x
```

```python
y
```

```python
# To find the correlation with x (input) and y(output ) data.It should be always near to + or -1
x.corrwith(y)
```

```python
##### Build the model
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x,y)
```

```python

regressor.score(x,y)
```

```python
# Allthree inputs are negatively correlated with output
regressor.coef_
```

```python
regressor.intercept_
```

```python
##### Predict on unseen data
```

```python
disp =221
hp = 102
wt = 2.91
```

```python
new =pd.DataFrame({
    'disp':[221],
    'hp':[102],
    'wt':[2.91]
})
```

```python
regressor.predict(new)
```
