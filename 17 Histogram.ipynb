```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
num = [np.random.randint(0,100) for x in range(1000)]
```

```python
print(num)
```

```python
plt.title('Historgram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(10,101,10))
plt.grid()
plt.hist(num, bins = range(10,101,10))
```

```python
img = cv2.imread('images/lenna.png',0)
plt.figure(figsize=(16,9))
plt.title('Image Historgram')
plt.xlabel('Values')
plt.ylabel('Count')
plt.xticks(range(10,256,10))
plt.grid()
plt.hist(img.flatten(), bins = range(10,256,10))
```
