```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the image
img = cv2.imread('images/lenna.png',0)
```

```python
new_img = img >150
inv_img = img <150
```

```python
plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('Original Image')
plt.imshow(img, cmap='gray')
plt.subplot(132)
plt.title('Binary Image')
plt.imshow(new_img, cmap='gray')
plt.subplot(133)
plt.title('Inv-Binary Image')
plt.imshow(inv_img, cmap='gray')
```
