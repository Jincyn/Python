```python
import cv2

#Load the image
image = cv2.imread("images/lenna.png")

#Flip the image horizondally
flipped_horizontal = cv2.flip(image,0)

#Flip the image Vertically
flipped_vertical = cv2.flip(image,1)

#Flip the image horizondally
flipped_diagonal = cv2.flip(image,-1)


#show the images
cv2.imshow('Original',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('Flipped Horizontal',flipped_horizontal)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
cv2.imshow('Flipped Vertical',flipped_vertical)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
cv2.imshow('Flipped Diagonal',flipped_diagonal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
flip_hor = cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB)
flip_ver = cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)
flip_dia = cv2.cvtColor(flipped_diagonal, cv2.COLOR_BGR2RGB)
img_org = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('Original Image')
plt.imshow(img_org)

plt.subplot(1,4,2)
plt.title('Horizontal Image')
plt.imshow(flip_hor)


plt.subplot(1,4,3)
plt.title('Vertical Image')
plt.imshow(flip_ver)

plt.subplot(1,4,4)
plt.title('Diagonal Image')
plt.imshow(flip_dia)

```
