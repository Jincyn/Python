```python
##### Feature Scaling
```

```python
# Take input data as Age and Estimated Salary
x = df[['Age','EstimatedSalary']]

# Take output data as Purchased
y = df['Purchased']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
##### Cross Validation
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,random_state=0)
```

```python
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_train
```

```python
##### Build the model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
# Train the Algorithm
```

```python
classifier.fit(x_train,y_train)
```

```python
# Predict on the unseen data
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred # here the salary  value is very high and age is very low scale. so normalisationis required
```

```python
# Problems in Probability Predictions
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
# Predict Probabilities
y_prob = classifier.predict_proba(x_test)[:,1]
```

```python
y_prob
```

```python
fpr, tpr, thresholds = roc_curve(y_test,y_prob)
roc_auc=roc_auc_score(y_test,y_prob)
```

```python
roc_auc
```

```python
#plot ROC curve
plt.plot(fpr,tpr,color='darkorange', lw=2,
         label=f'ROC Curve(area = {roc_auc:.2f})')
plt.plot([0,1],[0,1],color='navy', lw=2, linestyle='--')
plt.xlim([0.0,1.0])
plt.xlim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristics(ROC) curve')
plt.legend(loc='lower right')
```
