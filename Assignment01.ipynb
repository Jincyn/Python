```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/sachin.jpg')
```

```python
img.shape
```

```python
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
```

```python
r.shape, g.shape, b.shape
```

```python
plt.subplot(1,3,1)
plt.imshow(r,cmap='Reds')
plt.subplot(1,3,2)
plt.imshow(g,cmap='Greens')
plt.subplot(1,3,3)
plt.imshow(b,cmap='Blues')
```

```python
cv2.imshow('Red',r)
cv2.imshow('Green',g)
cv2.imshow('Blue',b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,4,1)
plt.title('gray')
plt.imshow(img,cmap='gray')
```
